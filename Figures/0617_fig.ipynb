```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import torch
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from torch.autograd import Variable
import torch.nn as nn
import pandas as pd


all_path="/content/drive/MyDrive/Colab_Notebooks/2023_MEP/1107_data/"



device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device to train")

mycase="mp"
casenum="14"
#-------------------------------------------------------------------------


if mycase[0:2]=="mp":
  path_Q=all_path+mycase+"_Q_"+casenum+".pkl"
  pkl_file = open(path_Q, 'rb')
  Q = pickle.load(pkl_file)
  pkl_file.close()
  Q_ten=torch.tensor(Q).to(device)

  dimension = 10 #dimension
  ad=0

  x_lb=-1.5
  x_ub=0.9
  y_lb=-0.5
  y_ub=1.9



  class NeuralNetwork1(nn.Module):
    def __init__(self,st,ed):
        super(NeuralNetwork1, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_tanh_stack = nn.Sequential(
            nn.Linear(1, 20),
            nn.Sigmoid(),
            nn.Linear(20, 20),
            nn.Sigmoid(),
            nn.Linear(20, 20),
            nn.Sigmoid(),
            nn.Linear(20, 20),
            nn.Sigmoid(),
            nn.Linear(20, 20),
            nn.Sigmoid(),
            nn.Linear(20, dimension+ad)
        )
        self.startpoint=Variable(torch.from_numpy(st),requires_grad=False).to(device)
        self.endpoint=Variable(torch.from_numpy(ed),requires_grad=False).to(device)


    def forward(self, s):
        s = self.flatten(s)
        x_pred = self.linear_tanh_stack(s)
        out=s*(1-s)*x_pred + (1-s)*self.startpoint + s*self.endpoint
        return out


  class NeuralNetwork2(nn.Module):
      def __init__(self,st,ed):
          super(NeuralNetwork2, self).__init__()
          self.flatten = nn.Flatten()

          self.stack_all=nn.ModuleList()
          for i in range(dimension+ad):
            self.stack_all.append(nn.Linear(1, 6))
            self.stack_all.append(nn.Sigmoid())
            for j in range(4):
              self.stack_all.append(nn.Linear(6, 6))
              self.stack_all.append(nn.Sigmoid())
            self.stack_all.append(nn.Linear(6, 1))

          self.startpoint=Variable(torch.from_numpy(st),requires_grad=False).to(device)
          self.endpoint=Variable(torch.from_numpy(ed),requires_grad=False).to(device)


      def forward(self, s):
          s = self.flatten(s)
          x_all=[]
          for i in range(dimension+ad):
            xx=self.stack_all[i*11](s)
            for j in range(10):
              xx=self.stack_all[i*11+j+1](xx)
            x_all.append(xx)
          x_pred = torch.cat(x_all,1)
          out=s*(1-s)*x_pred + (1-s)*self.startpoint + s*self.endpoint
          return out




  MEPpass_start_point=[6.23499399e-01,2.80377512e-02]+[0]*(dimension+ad-2)
  MEPpass_end_point=[-5.58223673e-01,1.44172580e+00]+[0]*(dimension+ad-2)

  MEPpass_start_point=np.array(MEPpass_start_point)
  MEPpass_end_point=np.array(MEPpass_end_point)

  MEPpass_start_point=np.matmul(Q,MEPpass_start_point.reshape(-1,1))
  MEPpass_end_point=np.matmul(Q,MEPpass_end_point.reshape(-1,1))

  MEPpass_start_point=MEPpass_start_point.reshape(1,-1)
  MEPpass_end_point=MEPpass_end_point.reshape(1,-1)

  def V_tensor(x):
    a = [ -1, -1, -6.5, 0.7]
    b = [0, 0, 11, 0.6]
    c = [ -10, -10, -6.5, 0.7]
    D = [ -200, -100, -170, 15]
    X = [1, 0, -0.5, -1]
    Y = [0, 0.5, 1.5, 1]
    sigma=0.5

    res=0
    for i in range(4):
      res=res+D[i]*torch.exp(a[i]*(x[0]-X[i])**2+b[i]*(x[0]-X[i])*(x[1]-Y[i])+c[i]*(x[1]-Y[i])**2)
    for i in range(dimension-2):
      res=res+1/(2*sigma**2)*x[i+2]**2
    return res




else:
  print("Check your code")


#=============================================================


def fig_countour(x_plot_prime,x_plot_prime_nobeta):
    global mycase
    #print("-------------------------------------------------------")
    plt.figure(figsize=(5,4))
    x = np.linspace(x_lb, x_ub,num=51,endpoint=True)
    y = np.linspace(y_lb, y_ub,num=51,endpoint=True)
    #print("x=",x)
    X,Y = np.meshgrid(x,y)
    X_new=X.reshape(-1,1)
    Y_new=Y.reshape(-1,1)
    XY=np.hstack((X_new,Y_new))
    for i in range(dimension-2):
      XY=np.hstack((XY,np.zeros(X_new.shape)))

    XY_tensor=torch.from_numpy(XY)
    X_list = []
    for i in range(dimension):
      X_list.append(XY_tensor[:, i:i+1].to(device))
    Z = V_tensor(X_list)

    Z_new=Z.reshape(X.shape).cpu().detach().numpy()

    plt.figure()
    CS = plt.contourf(X,Y,Z_new,50,cmap=mpl.cm.jet)
    plt.colorbar(CS)
    plt.plot(x_plot_prime_nobeta[:,0],x_plot_prime_nobeta[:,1],"r-",linewidth=2,markersize=3,label="Seperate networks")
    plt.plot(x_plot_prime[:,0],x_plot_prime[:,1],"c-",linewidth=2,markersize=3,label="FCN")

    plt.xlim((x_lb,x_ub))
    plt.ylim((y_lb,y_ub))
    plt.xlabel('$z_1$')
    plt.ylabel('$z_2$')
    plt.legend()
    plt.savefig("case"+mycase+"_"+casenum+"_"+"pathbothnets.pdf")

    if mycase=="mp":

      xx_save = x_plot
      save = pd.DataFrame(xx_save)
      save.to_csv(all_path+mycase+"_"+casenum+"_"+"path.csv",index=False,header=False)

      xx_save = x_plot_nobeta
      save = pd.DataFrame(xx_save)
      save.to_csv(all_path+mycase+"_"+casenum+"_"+"path_nobeta.csv",index=False,header=False)


      # output = open(all_path+mycase+"_"+casenum+"_"+"path.pkl", 'wb')
      # pickle.dump(x_plot,output)
      # output.close()
      # output = open(all_path+mycase+"_"+casenum+"_"+"path_nobeta.pkl", 'wb')
      # pickle.dump(x_plot_nobeta,output)
      # output.close()


#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

model1 = NeuralNetwork1(MEPpass_start_point,MEPpass_end_point).to(device)
model1 = model1.double()

model2 = NeuralNetwork2(MEPpass_start_point,MEPpass_end_point).to(device)
model2 = model2.double()



ss = np.linspace(0.001,0.999,40,endpoint=True)
ss = ss.reshape(-1,1)
ss = Variable(torch.from_numpy(ss),requires_grad=True).to(device)



#--------------prime---------------------

model_name= all_path+"/model_MEP_"+mycase+"_prime_stru_ori"+"_"+casenum+".pth"
model1.load_state_dict(torch.load(model_name))
x_prime = model1(ss)
if mycase=="mp":
  x_prime=torch.matmul(x_prime,Q_ten)
  x_prime=x_prime[:,0:2]
x_plot_prime=x_prime.cpu().detach().numpy()

#--------------prime_nobeta---------------------


model_name= all_path+"/model_MEP_"+mycase+"_prime_stru_10"+"_"+casenum+".pth"
model2.load_state_dict(torch.load(model_name))
x_prime_nobeta = model2(ss)
if mycase=="mp":
  x_prime_nobeta=torch.matmul(x_prime_nobeta,Q_ten)
  x_prime_nobeta=x_prime_nobeta[:,0:2]
x_plot_prime_nobeta=x_prime_nobeta.cpu().detach().numpy()

# #---------------initial----------------
# model_name =all_path+"model_MEP_"+mycase[0:2]+"_"+casenum+"pretrain"+".pth"
# model.load_state_dict(torch.load(model_name))
# x_initial = model(ss)
# if mycase=="mp":
#   x_initial=torch.matmul(x_initial,Q_ten)
#   x_initial=x_initial[:,0:2]
# x_plot_initial=x_initial.cpu().detach().numpy()




fig_countour(x_plot_prime,x_plot_prime_nobeta)








```
